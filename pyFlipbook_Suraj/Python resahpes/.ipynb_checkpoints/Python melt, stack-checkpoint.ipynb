## Stack and Unstack

### Stack Example
https://towardsdatascience.com/reshaping-pandas-dataframes-9812b3c1270e

Stack Kind of increases the index level i.e. (a) for simple df creates index-columns pairs (b) If dataframe has multi-level index, stack increases the index level

```python
df1 = pd.DataFrame({'city':['A','B','C'],
                   'day1':[22,25,28],
                   'day2':[10,14,13],})
                   #'day3':[25,22,26],
                   #'day4':[18,15,17],
                   #'day5':[12,14,18]})
df1
```

```python
day1_index = df1.set_index('day1')#.stack(level='day1')#.reset_index()
day1_index
```

```python
day1_index.stack()
```

```python
print(len(day1_index.stack().index.levels))
print(day1_index.stack().shape)
day1_index.stack().index[0]
```

```python
day1_index.stack().reset_index()
```

```python
print(day1_index.stack().reset_index().shape)
day1_index.stack().reset_index().index[3]
```

#### Dataframe with multi-level index

```python
import numpy as np
```

```python
tuples = [('A',1),('A',2),('A',3),('B',1),('A',2)]
print("Tuples looks like this : ")
print(tuples)
index = pd.MultiIndex.from_tuples(tuples, names=['first','second'])
print("\nIndex looks like this : ")
print(index)
df2 = pd.DataFrame(np.random.randint(10, size=(5,2)), 
                   index=index, columns=['column_x', 'column_y'])
print("\n\nAnd dataframe looks like this : ")
df2
```

##### If we apply stack function on this dataframe, the level of index will be increased

```python
print(len(df2.index.levels))
print(df2.index.levels)
df2.index.levels
```

```python
multi_level_df_stacked = df2.stack().to_frame()
multi_level_df_stacked
```

```python
print(len(multi_level_df_stacked.index.levels))
print(multi_level_df_stacked.index.levels)
multi_level_df_stacked.index.levels
```

### Unstack
It is the opposite of stack

```python
#multi_level_df_stacked.unstack()
```

```python
tuples = [('A',1),('A',2),('A',3),('B',1)]
print("Tuples looks like this : ")
print(tuples)
index = pd.MultiIndex.from_tuples(tuples, names=['first','second'])
print("\nIndex looks like this : ")
print(index)
df2 = pd.DataFrame(np.random.randint(10, size=(4,2)), 
                   index=index, columns=['column_x', 'column_y'])
print("\n\nAnd dataframe looks like this : ")
df2
```

```python
multi_level_df_stacked = df2.stack().to_frame()
multi_level_df_stacked
```

```python
multi_level_df_stacked.unstack()
```

## Pivot
It is used to explore the relationships among variables by allowing to represent data in different formats

```python
df = pd.DataFrame(
{"student": ["Andy", "Bernie", "Cindy", "Deb"],
"school": ["Z", "Y", "Z", "Y"],
"english": [10, 100, 1000, 10000], # eng grades
"math": [20, 200, 2000, 20000], # math grades
})
df
```

##### relationship between school and student

```python
df.pivot(index='school', columns='student')#.reset_index()
```

```python
df.pivot(index='school', columns='student', values='english')
```
